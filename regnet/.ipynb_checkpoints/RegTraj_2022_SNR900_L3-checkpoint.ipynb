```python
#-------------------------------------------------------#
#Initialize the Neural Network#
#-------------------------------------------------------#

# A) Define NN layer widths
#--------------------------
#Dimensions of hidden layers (from Liu abst.)
H1, H2, H3, H4 = 32, 256, 256, 32  

#The number of parameters you're aiming to predict
out_dim =  len(target_names) 

# B) Define NN hyperparameters
#-----------------------------
P = dict()


P['num_epoch']  = 15        # num. times entire training set passes through NN

P['lr']         = 0.0001    # learning rate for optimizer

P['batch_size'] = 50        # num. training samples in 1 iteration

if train_network:
    P['N_train']    = len(training_dataset)    # num. samples in training set

    P['N_validation'] = len(validation_dataset)# num. samples in validation set

P['N_test']     = len(testing_dataset)         # num. samples in testing set

# C) Fix the Random Seed
#-----------------------
seed = 657

np.random.seed(seed)

torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)

random.seed(seed)



# D) Get a Neural Network
#    (a) Initialize a Neural Network
#-------------------------------------
myNeuralNet = regularizationTrajNN(in_dim,         # Input layer size

                                   H1, H2, H3, H4, # Hidden layer sizes

                                   out_dim)        # Output layer size

myNeuralNet = myNeuralNet.to(device)


#    (b1) Load in the pretrained Neural Network
#---------------------------------------------
NeuralNetList = []

if not train_network:
    for Error in ErrorType:
        myNeuralNet.load_state_dict(torch.load(NET_PATH + Error+'_minMetric.pth'))
        #print(myNeuralNet)
        NeuralNetList.append(myNeuralNet)


#    (b2) Define loss function and optimizer if intializing a new Neural Network
if train_network:
    
    # (distance induced by Frobenius norm)**2/(target.numel())
    myLoss = nn.MSELoss() 

    myOptimizer = torch.optim.Adam(myNeuralNet.parameters(), lr=P['lr'])

#summarize the Neural Network
print(myNeuralNet)
```

```python
#-------------------------------------------------------#
#    Establish some Error Functions
#-------------------------------------------------------#

counter = 0
error_label_list = []

for Error in ErrorType:
    Error_Label = str(Error)
    
    if Error_Label == 'PE':
        PEnum = counter
    if Error_Label == 'SE':
        SEnum = counter
    if Error_Label == 'LPE':
        LPEnum = counter
    if Error_Label == 'LSE':
        LSEnum = counter
    
    counter = counter+1
    
    error_label_list.append(Error_Label)
    
def DefineErrorIndex(the_error):
    if the_error == 'PE':
        num = PEnum
    if the_error == 'SE':
        num = SEnum
    if the_error == 'LPE':
        num = LPEnum
    if the_error == 'LSE':
        num = LSEnum
    return num
```

```python
#-------------------------------------------------------#
#    Train the Neural Network
#-------------------------------------------------------#
if train_network:
        loss_e, validation_loss_e = np.zeros(P['num_epoch']), np.zeros(P['num_epoch'])
        
        this_metric_list = []
        best_metric_list = []
        best_epoch_list = []
        mean_error_e_list = []

        for Error in ErrorType:
            
            mean_error_e = torch.zeros((P['num_epoch'],out_dim), dtype=torch.float64)
            mean_error_e_list.append(mean_error_e)
            
            # Metric: For each parameter estimated, calculate average relative error. Then sum.
            this_metric_list.append(0.0)
            best_metric_list.append(1e10)  # large so it is immediately overwrriten.  
            best_epoch_list.append(0)
        
        print('this_metric_list',
        this_metric_list)
        print('best_metric_list',
        best_metric_list)
        print('best_epoch_list',
        best_epoch_list)
        print('error_label_list',
        error_label_list)
        print('mean_error_e_list',
        mean_error_e_list)
        
        #---
        num_batches = P['N_train']/P['batch_size']
        #===== ===== ===== ===== =====
        
        print('\n\n----  Starting Training  ----')

        print(f'Number of training samples: {P["N_train"]}')
        
        #---------------------------------------------------------
        #In an Epoch...
        #---------------------------------------------------------
        for epoch in range(P['num_epoch']):

            # a new epoch begins

            print(f'-- epoch {epoch}')

            running_loss = 0.0

            for batch_idx, (trajectories, targets) in enumerate(training_loader):

                trajectories, targets = trajectories.to(device), targets.to(device)

                # (trajectories, targets) is a mini-batch:

                #   -- trajectories is a Tensor of regularization trajectories of
                #        size P['batch_size']x(num_reg_params*num_trajectories)

                #   -- targets is a Tensor of target parameters of size
                #        P['batch_size']x(num_params)

                ####Apply Transforms here####
                ## Define Transform ##
                transform = "Shuffle"

                # 1) Initialize the gradient of the loss to zero

                myOptimizer.zero_grad()

                # 2) Compute the biexponential's parameters and their loss

                predictions = myNeuralNet(trajectories.float())

                loss = myLoss(predictions.double(), targets.double()) # (input, target)

                # 3) Estimate the loss' gradient with back propagation

                loss.backward()

                # 4) Update the NN weights

                myOptimizer.step()

                # 5) Estimate the overall loss over the mini-batch

                loss = loss.to("cpu")

                running_loss += loss.detach().numpy()

            loss_e[epoch] = running_loss/num_batches # ~ (total Fro-error)/N_train

            with torch.no_grad():

                for trajectories, targets in validation_loader:
                    trajectories, targets = trajectories.to(device), targets.to(device)

                    predictions = myNeuralNet(trajectories.float())

                    validation_loss_e[epoch] = myLoss(predictions.double(), targets.double())
                    
                    
                    for Error in ErrorType:
                        Error_Label = str(Error)
                        if Error == 'PE':
                            num = PEnum
                            mean_error_e_list[num][epoch,:] = em.mean_percent_error(predictions, targets)
                        if Error == 'SE':
                            num = SEnum
                            mean_error_e_list[num][epoch,:] = em.mean_square_error(predictions, targets)
                        if Error == 'LPE':
                            num = LPEnum
                            mean_error_e_list[num][epoch,:] = em.log_mean_percent_error(predictions, targets)
                        if Error == 'LSE':
                            num = LSEnum
                            mean_error_e_list[num][epoch,:] = em.log_mean_square_error(predictions, targets)
                            
                        
                        this_metric_list[num] = torch.sum(mean_error_e_list[num][epoch,:], dim=0)

                        if this_metric_list[num] < best_metric_list[num]:

                            best_metric_list[num], best_epoch_list[num] = this_metric_list[num], epoch
                            
                            torch.save(myNeuralNet.state_dict(), NET_PATH+Error_Label+'_minMetric.pth')
            '''
            #debug
            print('this_metric_list', this_metric_list)
            print('best_metric_list', best_metric_list)
            print('best_epoch_list', best_epoch_list)
            print('error_label_list', error_label_list)
            print('mean_error_e_list', mean_error_e_list)
            '''
            
            for Error in ErrorType:
                if Error == 'PE':
                    num = PEnum
                if Error == 'SE':
                    num = SEnum
                if Error == 'LPE':
                    num = LPEnum
                if Error == 'LSE':
                    num = LSEnum
                print(f'     Average Loss: {loss_e[epoch]:.2e}  |  Average'+error_label_list[num]+' error (%) \n'+

                      f'T21: {mean_error_e_list[num][epoch,0]:.3f} | '+ #changed 1 to 0

                      f'T22: {mean_error_e_list[num][epoch,1]:.3f} | '+ #changed 2 to 1

                      f'Sum: {this_metric_list[num]:.4f}')
                
#--------------------------------------------------------------------------
                
        epoch_dynamics = {'epoch': torch.Tensor(range(P['num_epoch'])),

                          'loss': loss_e,

                          'validation_loss': validation_loss_e,

                          error_label_list[num]+'error_mean_rel_error': mean_error_e_list[num],

                          'best_epoch': best_epoch_list[num],

                          'min_sum_mean_rel_error': best_metric_list[num]}

        torch.save(epoch_dynamics, FIG_PATH+'epoch_data.pth')


        torch.save(myNeuralNet.state_dict(), NET_PATH+'final_epoch'+'.pth')
```

```python
#-------------------------------------------------------#
#    Test the Neural Network
#-------------------------------------------------------#

#For saving some essential parameters
masked_abs_error_nlls_list = []
sum_error_nlls_list = []

mean_T21_error_nlls_list = []
mean_T22_error_nlls_list = []

mean_sum_error_nlls_list = []

mean_T21_abs_error_nlls_list = []
mean_T22_abs_error_nlls_list= []

#For saving the masked parameter outputs for various errors
masked_params_nlls_list = []
masked_params_test_list = []
masked_noisy_decays_list = []
masked_error_nlls_list = []
masked_trajectory_list = []

#For saving the T21 and T22 computated data
predicted_T21_means_nlls_list = []
predicted_T21_stdevs_nlls_list = []
T21_error_means_nlls_list = []   #abs_
T21_error_stdevs_nlls_list = []  #abs_

twoT21_error_means_nlls_list = []  #PE SE LPE LSE
twoT21_error_stdevs_nlls_list = [] #PE SE LPE LSE

predicted_T22_means_nlls_list = []
predicted_T22_stdevs_nlls_list = []
T22_error_means_nlls_list = []   #abs_
T22_error_stdevs_nlls_list = []  #abs_

twoT22_error_means_nlls_list = []  #PE SE LPE LSE
twoT22_error_stdevs_nlls_list = [] #PE SE LPE LSE

with torch.no_grad(): # indicates to auto_grad not to update computational graph

    for trajectory, params_test in testing_loader:
        
        trajectory, params_test = trajectory.to(device), params_test.to(device)
        
        print('\n\n----  Starting Testing  ----')

        print(f'    -- Number of tests is: {P["N_test"]}')
        
        ##---------------------------------##

        ##    Evaluate NLLS Performance    ##

        ##---------------------------------##

        print('    -- Evaluating NLLS performance...')

        # A) Initialize

        params_nlls = torch.zeros((P['N_test'], out_dim), dtype=torch.float64, device= device)

        noisy_decays= torch.zeros((P['N_test'], num_times), dtype=torch.float64)

        for sample in range(P['N_test']):

            noisy_decays[sample,:] = testing_dataset.master_frame['noisy_decay'][sample]

        #params_nlls[:,0] = torch.Tensor(testing_dataset.master_frame['c1_nlls'])

        params_nlls[:,0] = torch.Tensor(testing_dataset.master_frame['T21_nlls']) # changed 1 to 0

        params_nlls[:,1] = torch.Tensor(testing_dataset.master_frame['T22_nlls']) # changed 2 to 1

        print(params_nlls)
        
        params_nlls = params_nlls.cpu()
        
        params_nlls = parameter_swap_same_tensor_two_variables(params_nlls)
        
        params_nlls = params_nlls.to(device)

        # B) Calculate absolute error

        abs_error_nlls = torch.abs(params_nlls - params_test)
        
        # C) Calculate errors.
        nlls_perc_error_denom = params_test.clone()
        
        # D) Discard all samples that contain outliers. A parameter estimate is an...
        #      ...outlier if it is less than Q1 - 1.5*IQR or greater than Q3 + 1.5*IQR
        # mask based on T21 percent errors
        
        for Error in ErrorType:
            if Error == 'PE':
                num = PEnum
                error_nlls = 100.0*abs_error_nlls/nlls_perc_error_denom
            if Error == 'SE':
                num = SEnum
                error_nlls = (params_nlls - params_test)**2
            if Error == 'LPE':
                num = LPEnum
                error_nlls = 100.0*torch.log(abs_error_nlls)/torch.log(nlls_perc_error_denom)
            if Error == 'LSE':
                num = LSEnum
                error_nlls = (torch.log(params_nlls) - torch.log(params_test))**2
            
            #===== ===== =====
            #     MASK 1
            #===== ===== =====
            #---
            error_nlls = error_nlls.to('cpu')
            mask1 = em.mask_outliers(error_nlls[:,0], TOL)
            error_nlls = error_nlls.to(device)
            #---
            params_nlls_m1 = params_nlls[mask1,:]
            params_test_m1 = params_test[mask1,:]
            noisy_decays_m1 = noisy_decays[mask1,:]
            error_nlls_m1 = error_nlls[mask1,:]
            trajectory_m1 = trajectory[mask1,:]
            
            # mask based on T22 percent errors
            error_nlls_m1 = error_nlls_m1.to('cpu')
            
            #===== ===== =====
            #     MASK 2
            #===== ===== =====
            #---
            error_nlls_m1 = error_nlls_m1.to('cpu')
            mask2 = em.mask_outliers(error_nlls_m1[:,1], TOL)
            error_nlls_m1 = error_nlls_m1.to(device)
            #---
            masked_params_nlls = params_nlls_m1[mask2,:]
            masked_params_test = params_test_m1[mask2,:]
            masked_noisy_decays = noisy_decays_m1[mask2,:]
            masked_error_nlls = error_nlls_m1[mask2,:]
            masked_trajectory = trajectory_m1[mask2,:]
            #- - - - - - - - - - - - - - - - - - - - - - - - - - -
            #                    Saving...
            masked_params_nlls_list.append(masked_params_nlls)
            masked_params_test_list.append(masked_params_test)
            print('debug', masked_params_test)
            masked_noisy_decays_list.append(masked_noisy_decays)
            masked_error_nlls_list.append(masked_error_nlls)
            masked_trajectory_list.append(masked_trajectory)
            #- - - - - - - - - - - - - - - - - - - - - - - - - - -
            
            #print('mask2', mask2)
            #print('params_nlls_m1', params_nlls_m1)
            #print('masked_params_nlls', masked_params_nlls)
            
            print(f'        -> Discarded {P["N_test"]-len(masked_params_nlls[:,0])} outliers of {P["N_test"]} samples')
            
            discard = P["N_test"]-len(masked_params_nlls[:,0])
            
            discard_list.append(discard)
            #[@@@]  Track discarded samples    P["N_test"]-len(masked_params_nlls[:,0])
            
            # Calculate sample-averaged percent error for each parameter (no outliers)

            mean_T21_error_nlls, mean_T22_error_nlls = torch.mean(masked_error_nlls, dim=0)
            mean_T21_error_nlls_list.append(mean_T21_error_nlls)
            mean_T22_error_nlls_list.append(mean_T22_error_nlls)
            
            # Calculate absolute errors (no outliers)
            masked_abs_error_nlls = torch.abs(masked_params_nlls - masked_params_test)
            masked_abs_error_nlls_list.append(masked_abs_error_nlls)
            
            mean_abs_error_nlls = torch.mean(masked_abs_error_nlls, dim=0)
            mean_T21_abs_error_nlls, mean_T22_abs_error_nlls = mean_abs_error_nlls
            
            #  Save...
            mean_T21_abs_error_nlls_list.append(mean_T21_abs_error_nlls)
            mean_T22_abs_error_nlls_list.append(mean_T22_abs_error_nlls)
            
            # Sum of errors over parameters
            sum_error_nlls = torch.sum(masked_error_nlls, dim=1)
            sum_error_nlls_list.append(sum_error_nlls)
            
            # Average over samples
            mean_sum_error_nlls = torch.mean(sum_error_nlls, dim=0)
            mean_sum_error_nlls_list.append(mean_sum_error_nlls)
            
            if Error == 'PE':
                (_, predicted_T21_means_nlls, 
                 predicted_T21_stdevs_nlls, 
                 T21_error_means_nlls, 
                 T21_error_stdevs_nlls, 
                 twoT21_error_means_nlls, 
                 twoT21_error_stdevs_nlls) = em.PE_prediction_averages_stdevs(em, 
                                                                            masked_params_test[:,0], 
                                                                            masked_params_nlls[:,0])
                (_, predicted_T22_means_nlls, 
                 predicted_T22_stdevs_nlls, 
                 T22_error_means_nlls, 
                 T22_error_stdevs_nlls, 
                 twoT22_error_means_nlls, 
                 twoT22_error_stdevs_nlls) = em.PE_prediction_averages_stdevs(em, 
                                                                            masked_params_test[:,1], 
                                                                            masked_params_nlls[:,1])
            if Error == 'SE':
                (_, predicted_T21_means_nlls, 
                 predicted_T21_stdevs_nlls, 
                 T21_error_means_nlls, 
                 T21_error_stdevs_nlls, 
                 twoT21_error_means_nlls, 
                 twoT21_error_stdevs_nlls) = em.SE_prediction_averages_stdevs(em, 
                                                                            masked_params_test[:,0], 
                                                                            masked_params_nlls[:,0])
                (_, predicted_T22_means_nlls, 
                 predicted_T22_stdevs_nlls, 
                 T22_error_means_nlls, 
                 T22_error_stdevs_nlls, 
                 twoT22_error_means_nlls, 
                 twoT22_error_stdevs_nlls) = em.SE_prediction_averages_stdevs(em, 
                                                                            masked_params_test[:,1], 
                                                                            masked_params_nlls[:,1])
            if Error == 'LPE':
                (_, predicted_T21_means_nlls, 
                 predicted_T21_stdevs_nlls, 
                 T21_error_means_nlls, 
                 T21_error_stdevs_nlls, 
                 twoT21_error_means_nlls, 
                 twoT21_error_stdevs_nlls) = em.LPE_prediction_averages_stdevs(em, 
                                                                            masked_params_test[:,0], 
                                                                            masked_params_nlls[:,0])
                (_, predicted_T22_means_nlls, 
                 predicted_T22_stdevs_nlls, 
                 T22_error_means_nlls, 
                 T22_error_stdevs_nlls, 
                 twoT22_error_means_nlls, 
                 twoT22_error_stdevs_nlls) = em.LPE_prediction_averages_stdevs(em, 
                                                                            masked_params_test[:,1], 
                                                                            masked_params_nlls[:,1])
            if Error == 'LSE':
                (_, predicted_T21_means_nlls, 
                 predicted_T21_stdevs_nlls, 
                 T21_error_means_nlls, 
                 T21_error_stdevs_nlls, 
                 twoT21_error_means_nlls, 
                 twoT21_error_stdevs_nlls) = em.LSE_prediction_averages_stdevs(em, 
                                                                            masked_params_test[:,0], 
                                                                            masked_params_nlls[:,0])
                (_, predicted_T22_means_nlls, 
                 predicted_T22_stdevs_nlls, 
                 T22_error_means_nlls, 
                 T22_error_stdevs_nlls, 
                 twoT22_error_means_nlls, 
                 twoT22_error_stdevs_nlls) = em.LSE_prediction_averages_stdevs(em, 
                                                                            masked_params_test[:,1], 
                                                                            masked_params_nlls[:,1])
            
            predicted_T21_means_nlls_list.append(predicted_T21_means_nlls)
            predicted_T21_stdevs_nlls_list.append(predicted_T21_stdevs_nlls)
            T21_error_means_nlls_list.append(T21_error_means_nlls)
            T21_error_stdevs_nlls_list.append(T21_error_stdevs_nlls)
            twoT21_error_means_nlls_list.append(twoT21_error_means_nlls)
            twoT21_error_stdevs_nlls_list.append(twoT21_error_stdevs_nlls)
            
            predicted_T22_means_nlls_list.append(predicted_T22_means_nlls)
            predicted_T22_stdevs_nlls_list.append(predicted_T22_stdevs_nlls)
            T22_error_means_nlls_list.append(T22_error_means_nlls)
            T22_error_stdevs_nlls_list.append(T22_error_stdevs_nlls)
            
            twoT22_error_means_nlls_list.append(twoT22_error_means_nlls)
            twoT22_error_stdevs_nlls_list.append(twoT22_error_stdevs_nlls)
            
#print('predicted_T21_means_nlls_list', predicted_T21_means_nlls_list)
#print('predicted_T21_stdevs_nlls_list', predicted_T21_stdevs_nlls_list)
#print('T21_error_means_nlls_list', T21_error_means_nlls_list)
#print('T21_error_stdevs_nlls_list', T21_error_stdevs_nlls_list)
#print('twoT21_error_means_nlls_list', twoT21_error_means_nlls_list)
#print('twoT21_error_stdevs_nlls_list', twoT21_error_stdevs_nlls_list)
```

```python
#-------------------------------------------------------#
#     Find the sampels with the Worst Errors
#-------------------------------------------------------#

# C) Find the samples with most error and reconstruct their signals

#------------------------------------------------------------------

## C.1.1) Find the worst offenders for the NN for all types of error

worst_T21_error_list = []
worst_T21_idx_list = []
worst_T22_error_list = []
worst_T22_idx_list = []
worst_sum_error_list = []
worst_sum_err_idx_list = []

signal_worst_T21_true_list = []
signal_worst_T21_nlls_list = []
signal_worst_T21_nn_list = []

signal_worst_T22_true_list = []
signal_worst_T22_nlls_list = []
signal_worst_T22_nn_list = []

worst_T21_error_nlls_list = []
worst_T21_idx_nlls_list = []
worst_T22_error_nlls_list = []
worst_T22_idx_nlls_list = []
worst_sum_error_nlls_list = []
worst_sum_err_idx_nlls_list = []

#Print the Parameters (to preface C.2)
reg_parameters = testing_dataset.master_frame['reg_params'][0]
print(reg_parameters)
method_label, abr = methodLabel(INPUT_TYPE, num_copies=num_copies, num_noisy_copies=num_noisy_copies, decay_input=decay_inputs, regularization_parameters=reg_parameters)

#Establish Time Bounds (to preface C.3)
# Initialize signal acquisition times
time = torch.linspace(time_bounds[0], time_bounds[1], num_times,
                      dtype=torch.float64, device = device)


for Error in ErrorType:
    if Error == 'PE':
        num = PEnum
        error_nlls = 100.0*abs_error_nlls/nlls_perc_error_denom
    if Error == 'SE':
        num = SEnum
        error_nlls = (params_nlls - params_test)**2
    if Error == 'LPE':
        num = LPEnum
        error_nlls = 100.0*torch.log(abs_error_nlls)/torch.log(nlls_perc_error_denom)
    if Error == 'LSE':
        num = LSEnum
        error_nlls = (torch.log(params_nlls) - torch.log(params_test))**2
    
    # For the NN
    # - - -
    worst_T21_error, worst_T21_idx = torch.max(abs_error_list[num][:,0],0)
    worst_T22_error, worst_T22_idx = torch.max(abs_error_list[num][:,1],0)
    worst_sum_error, worst_sum_err_idx = torch.max(sum_error_list[num],0)
    
    worst_T21_error_list.append(worst_T21_error)
    worst_T21_idx_list.append(worst_T21_idx)
    worst_T22_error_list.append(worst_T22_error)
    worst_T22_idx_list.append(worst_T22_idx)
    worst_sum_error_list.append(worst_sum_error)
    worst_sum_err_idx_list.append(worst_sum_err_idx)
    
    # Repeat for NLLS
    # - - -
    worst_T21_error_nlls, _ = torch.max(masked_abs_error_nlls_list[num][:,0],0)
    worst_T22_error_nlls, _ = torch.max(masked_abs_error_nlls_list[num][:,1],0)
    worst_sum_error_nlls, _ = torch.max(sum_error_nlls_list[num],0)
    
    worst_T21_error_nlls_list.append(worst_T21_error_nlls)
    #worst_T21_idx_nlls_list.append(worst_T21_idx_nlls)
    worst_T22_error_nlls_list.append(worst_T22_error_nlls)
    #worst_T22_idx_nlls_list.append(worst_T22_idx_nlls)
    worst_sum_error_nlls_list.append(worst_sum_error_nlls)
    #worst_sum_err_idx_nlls_list.append(worst_sum_err_idx_nlls)
    
    ## C.2) Print summary of errors
    
    #[*************************]
    
    print('Nonlinear Least Squares (NLLS):\n-------------------------------')

    print(f'        |  Max abs. error | Average abs. error | Average '+error_label_list[num]+' error')

    print(f'    T21 | {worst_T21_error_nlls:03.3f} | {mean_T21_abs_error_nlls:03.3f} | {mean_T21_error_nlls:.3f}')

    print(f'    T22 | {worst_T22_error_nlls:03.3f} | {mean_T22_abs_error_nlls:03.3f} | {mean_T22_error_nlls:.3f}')
    
    print(method_label+' Neural Net ('+abr+'):\n----------------')

    print(f'        |  Max abs. error | Average abs. error | Average '+error_label_list[num]+'. error')

    print(f'    T21 | {worst_T21_error:03.3f} | {mean_T21_abs_error:03.3f} | {mean_T21_error:.3f}')

    print(f'    T22 | {worst_T22_error:03.3f} | {mean_T22_abs_error:03.3f} | {mean_T22_error:.3f}')
    
    print(f'\nNeural Net Performance Metrics\n------------------------------')

    print(f'    -- max over test samples of sum of '+error_label_list[num]+' errors ('+abr+f' v.s. NLLS): {worst_sum_error:.3f} v.s. {worst_sum_error_nlls:.3f}')

    print(f'    -- test-averaged sum of percent errors ('+abr+f' v.s. NLLS): {mean_sum_error_list[num]:.3f} v.s. {mean_sum_error_nlls_list[num]:.3f}')
    
    if train_network:

        print('Best Epoch',error_label_list[num], best_metric_list[num].item())
        print('Best Epoch Sum of Average',error_label_list[num],'errors',best_metric_list[num].item())
              
    time = time.to('cpu')
      
    masked_params_test_list[num] = masked_params_test_list[num].to('cpu')
    masked_params_nlls_list[num] = masked_params_nlls_list[num].to('cpu')
    params_nn_list[num] = params_nn_list[num].to('cpu')
              
    ## C.3) Reconstruct signals for worst T2,1 estimation
              
    signal_worst_T21_true = myTrueModel_2param(time, c1, c2, masked_params_test_list[num][worst_T21_idx,0],
                                    masked_params_test_list[num][worst_T21_idx,1], signalType=mySignalType)
    
    signal_worst_T21_true_list.append(signal_worst_T21_true)
    
    signal_worst_T21_nlls = myTrueModel_2param(time, c1, c2, masked_params_nlls_list[num][worst_T21_idx,0],
                                        masked_params_nlls_list[num][worst_T21_idx,1], signalType=mySignalType)

    signal_worst_T21_nlls_list.append(signal_worst_T21_nlls)
    
    signal_worst_T21_nn = myTrueModel_2param(time, c1, c2, params_nn_list[num][worst_T21_idx,0],
                                      params_nn_list[num][worst_T21_idx,1],signalType=mySignalType)
    
    signal_worst_T21_nn_list.append(signal_worst_T21_nn)
    
    ## C.4) Reconstruct signals for worst T2,2 estimation
              
    signal_worst_T22_true = myTrueModel_2param(time, c1, c2, masked_params_test_list[num][worst_T22_idx,0],
                                    masked_params_test_list[num][worst_T22_idx,1], signalType=mySignalType)

    signal_worst_T22_true_list.append(signal_worst_T22_true)
    
    signal_worst_T22_nlls = myTrueModel_2param(time, c1, c2, masked_params_nlls_list[num][worst_T22_idx,0],
                                        masked_params_nlls_list[num][worst_T22_idx,1], signalType=mySignalType)
    signal_worst_T22_nlls_list.append(signal_worst_T22_nlls)
    
    signal_worst_T22_nn = myTrueModel_2param(time, c1, c2, params_nn_list[num][worst_T22_idx,0],
                                      params_nn_list[num][worst_T22_idx,1],signalType=mySignalType)
    signal_worst_T22_nn_list.append(signal_worst_T22_nn)
    
    ## C.6) Reconstruct signals for worst sum relative error sample
    signal_worst_sum_rel_err_true = myTrueModel_2param(time, c1, c2, masked_params_test_list[num][worst_sum_err_idx,0],
                                            masked_params_test_list[num][worst_sum_err_idx,1],signalType=mySignalType)

    signal_worst_sum_rel_err_nlls = myTrueModel_2param(time, c1, c2, masked_params_nlls_list[num][worst_sum_err_idx,0],
                                                masked_params_nlls_list[num][worst_sum_err_idx,1],signalType=mySignalType)

    signal_worst_sum_rel_err_nn = myTrueModel_2param(time, c1, c2, params_nn_list[num][worst_sum_err_idx,0],
                                              params_nn_list[num][worst_sum_err_idx,1],signalType=mySignalType)
              
    time = time.to(device)
    masked_params_test_list[num] = masked_params_test_list[num].to(device)
    masked_params_nlls_list[num]= masked_params_nlls_list[num].to(device)
    params_nn_list[num] = params_nn_list[num].to(device)
```
